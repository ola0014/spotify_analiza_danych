Najpopularniejsze piosenki w całej twórczości

```python
new2 = filtered_tracks.sort_values(by=['popularity'], ascending=False).head(10)
new2
```

```python
import matplotlib.colors as mcolors
```

```python
# Tworzenie gradientu kolorów
gradient = np.linspace(0, 1, len(new2['name']))

# Tworzenie ombre kolorów dla słupków
colors = [mcolors.to_hex(plt.cm.RdPu(grad)) for grad in gradient]

new2['label'] = new2['name'] + ' (' + new2['year'].astype(str) + ')'

new2.plot.barh(x='label', y='popularity', color=colors, legend=None)
plt.xlabel('popularity')
plt.title('Najpopularniejsze piosenki The Weekndu', fontsize=12)
```
